```python
import numpy as np
import glob
import pandas as pd
```

```python
file_pattern = "logs/wide_*_select_scan.perf"
all_files = glob.glob(file_pattern)
```

```python
all_files
```

```python
df_list = []
for file_path in all_files:
    temp_df = pd.read_csv(file_path)
    temp_df['source_file'] = file_path
    df_list.append(temp_df)
```

```python
df_all = pd.concat(df_list, ignore_index=True)
```

```python
df_all['ipc'] = df_all['instructions'] / df_all['cpu_cycles']
df_all['llc_miss_rate'] = df_all['llc_misses'] / df_all['llc_references']
df_all['latency_ms'] = df_all['latency'] / 1e6
```

```python
df_all
```

```python
metrics = ['ipc', 'llc_miss_rate', 'latency_ms', 'l1d_read_miss']
df_all.groupby('source_file')[metrics].std() / df_all.groupby('source_file')[metrics].mean() * 100 
```

```python
summary = df_all.groupby('source_file')[metrics].mean()
```
